```python
import time
sales={}

no=int(input("Enter how many a number of products you have to sale: "))


for i in range(1,no+1):
   no= int(input("Enter a sale number : "))
   const_name = input("Enter Constomer Name :  ")
   ui_prod=str(input("Enter a product id: "))
   ui_quant=int(input("Enter a quantity number: "))
  
   if record[ui_prod]['qn'] >= ui_quant :
       print("*******************************") 
       print("Product Name: ",record[ui_prod]['name'])
       print("Product id : ",ui_prod)
       print("Per Product Price  : ",record[ui_prod]['pr'])
       print("Product Company Name : ",record[ui_prod]['company_name'])
       print("Product Manufactured date : ",record[ui_prod]['Mfg.date'])
       print("*******************************") 
       print ("Billing Amount : ",record[ui_prod]['pr']*ui_quant)

       sales[no]={'Customer Name':const_name, 'prod': ui_prod, 'qn': ui_quant, 'amount': record[ui_prod]['pr']*ui_quant, 'name': record[ui_prod]['name'], 'time': time.ctime()}
   
    
       record[ui_prod]['qn']=record[ui_prod]['qn'] - ui_quant
   else:
      print("sorry,That much qunatity number is not available now\n")   
    
print("Total sales made today")


```
